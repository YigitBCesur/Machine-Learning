# III-Missing Value Analysis

```python
import numpy as np
import pandas as pd
```

## Hızlı Çözüm

```python
# Create a dataset with 40 rows and 5 columns
data = {
    'A': np.random.randint(1, 11, 40),
    'B': np.random.choice(['X', 'Y', 'Z', 'W'], 40),
    'C': np.random.uniform(0, 1, 40),
    'D': np.random.choice([1, 2, 3, np.nan], 40),
    'E': np.random.choice([True, False, np.nan], 40)
}

df = pd.DataFrame(data)

# Introduce missing values in 20% of the dataset
missing_indices = np.random.choice(range(40), size=8, replace=False)
for col in df.columns:
    df.loc[missing_indices, col] = np.nan

print(df)
```
