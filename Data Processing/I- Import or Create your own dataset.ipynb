### **Create 100 rows + 5 columns ( numerical + categorical variables )dataset**

```python
import pandas as pd
import random

# Define the US States and car_purchased categories
us_states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"]

graduate_categories = ["High School", "Middle School", "Bachelor's", "Master's", "Doctorate"]
car_purchased_categories = ["Yes", "No"]

# Create a list to store the data
data = []

# Generate random data for the dataset
for _ in range(100):
    state = random.choice(us_states)
    age = random.randint(18, 65)
    salary = round(random.uniform(30000, 100000), -3)  # Round to the nearest thousand
    graduate = random.choice(graduate_categories)
    car_purchased = random.choice(car_purchased_categories)
    data.append([state, age, salary, graduate, car_purchased])

# Create a DataFrame
purchased = pd.DataFrame(data, columns=["US States", "Ages", "Salary", "Graduate", "Car Purchased"])


# Print the first few rows of the dataset
print(purchased.head())

# saving dataset
purchased.to_csv('purchased.csv')
```

### **"HOUSES"  dataset :**

we generate a dataset with 300 houses. We introduce outliers for the "price" column and missing values for the "square_feet" and "year_built" columns. You can further customize the data generation process, outlier rates, and missing value rates to meet your specific needs

```python
# Define ordinal categories for crime_zone
crime_zone_categories = ["Low", "Middle", "High"]

# Define features
data = []

for _ in range(300):
    location = f"Location_{random.randint(1, 10)}"
    bedrooms = random.randint(1, 6)
    bathrooms = random.uniform(1, 4)
    crime_zone = random.choice(crime_zone_categories)
    swimming_pool = random.choice(["Yes", "No"])
    square_feet = random.randint(1000, 4000)
    year_built = random.randint(1960, 2023)
    has_basement = random.choice(["Yes", "No"])
    has_garage = random.choice(["Yes", "No"])
    price = random.randint(100000, 1000000)

    # Introduce missing values randomly
    if random.random() < 0.02:
        square_feet = np.nan
    if random.random() < 0.04:
        year_built = np.nan
    if random.random() < 0.05:
        bedrooms = np.nan
    if random.random() < 0.02:
        bathrooms = np.nan
    if random.random() < 0.05:
        has_basement = np.nan
    if random.random() < 0.05:
        has_garage = np.nan
    if random.random() < 0.02:
        swimming_pool = np.nan
    if random.random() < 0.03:
        crime_zone = np.nan


    data.append([location, bedrooms, bathrooms, crime_zone, swimming_pool, square_feet, year_built, has_basement, has_garage, price])

# Create a DataFrame
houses = pd.DataFrame(data, columns=["Location", "Bedrooms", "Bathrooms", "Crime Zone", "Swimming Pool", "Square Feet", "Year Built", "Has Basement", "Has Garage", "Price"])

# Print the first few rows of the dataset
print(houses.head())

# saving dataset
houses.to_csv('houses.csv')
```

### **"SALES" dataset :**
Creating a synthetic **Sales** dataset with 1000 rows and 9 columns including missing values, outliers, and a mix of numeric and ordinal-nominal values

```python
# Set a random seed for reproducibility
np.random.seed(0)

# Generate data for the sales dataset
n = 1000

data = {
    'OrderID': range(1, n + 1),
    'Product': np.random.choice(['Product A', 'Product B', 'Product C', 'Product D'], n),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Home Appliances', 'Toys'], n),
    'Quantity': np.random.randint(1, 100, n),
    'Price': np.random.uniform(10, 500, n),
    'Discount': np.random.choice([0, 0.05, 0.1, 0.15], n),
    'Customer Rating': np.random.uniform(1, 5, n),
    'Delivery Time (days)': np.random.randint(1, 10, n),
    'Sales Rep': np.random.choice(['John', 'Sarah', 'Mike', 'Emily'], n),
}

# Introduce missing values
for column in data:
    missing_indices = np.random.choice(range(n), size=n//10, replace=False)
    data[column] = np.where(np.isin(range(n), missing_indices), np.nan, data[column])

# Introduce outliers
data['Price'][data['Price'] > 450] = data['Price'][data['Price'] > 450] * 2
data['Quantity'][data['Quantity'] > 80] = data['Quantity'][data['Quantity'] > 80] * 2

# Create the DataFrame
sales = pd.DataFrame(data)

# Display the first few rows of the dataset
print(sales.head())

# Save the dataset to a CSV file
sales.to_csv('sales.csv', index=False)
```
