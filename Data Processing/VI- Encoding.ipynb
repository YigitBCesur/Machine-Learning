## Variable Encoding

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
distance = np.random.randint(2, 13, size=30)
sex = np.random.choice(["female", "male"], size=30)
weight = np.random.randint(100, 221, size=30)
run_times = np.random.choice(["morning", "afternoon", "evening"], size=30)

# Create a DataFrame from the data
df = pd.DataFrame({
    "distance": distance,
    "sex": sex,
    "weight": weight,
    "run_time": run_times
})


df.head()
```

### 0-1 Transformation

```python
from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["new_sex"] = lbe.fit_transform(df["sex"])
```

```python
df
```

### "1 ve others (0) " encoding

```python
df.head()
```

```python
df["run_time"].str.contains("afternoon")
```

```python
df["new_times"] = np.where(df["run_time"].str.contains("afternoon"), 1, 0)
```
